```python
def create_playlist(playlist_name):
    user_id = sp.me()["id"]  # Get the current user's Spotify ID
    playlist = sp.user_playlist_create(user_id, playlist_name)  # Create a new playlist
    return playlist
```

```python
def add_songs_to_playlist(playlist_id, song_list):
    for song in song_list:
        results = sp.search(q=f"track:{song}", type="track", limit=1)
        if results["tracks"]["items"]:
            track_uri = results["tracks"]["items"][0]["uri"]
            sp.playlist_add_items(playlist_id, [track_uri])
#         else:
        else:
        
            print(f"Song '{song}' not found on Spotify.")
```
